```python
import os
import pandas as pd

# Verzeichnis mit den Dateien
dataset_path = "/Users/nicko/.cache/kagglehub/datasets/whisperingkahuna/premier-league-2324-team-and-player-insights/versions/5"

# Alle Dateien im Verzeichnis auflisten
files = os.listdir(dataset_path)
print("Files in dataset directory:", files)

# Dictionary für die DataFrames
dataframes = {}

# Jede CSV-Datei laden und im Dictionary speichern
for file in files:
    if file.endswith(".csv"):  # Nur .csv-Dateien verarbeiten
        file_path = os.path.join(dataset_path, file)
        dataframe_name = file.replace(".csv", "")  # Name ohne ".csv"
        dataframes[dataframe_name] = pd.read_csv(file_path)
        print(f"Loaded {file} as DataFrame with name '{dataframe_name}'.")

# Beispielzugriff auf einen DataFrame
print(dataframes.keys())  # Zeigt die Namen der gespeicherten DataFrames
print(dataframes['matches_23_24'].head())  # Beispiel: Zeigt die ersten Zeilen eines spezifischen DataFrames
```
